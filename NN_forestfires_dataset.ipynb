```python
from sklearn.preprocessing import LabelEncoder
```

```python
label_encoder= LabelEncoder()
forest.size_category=label_encoder.fit_transform(forest.size_category)
```

```python
forest.size_category
```

```python
x = forest.iloc[:,2:30]
y = forest.iloc[:,30]
```

## We need to do standardization of data

```python
array = x.values
scalar = StandardScaler().fit(array)
X_standardized = scalar.transform(array)
```

```python
X_standardized.shape
```

## Creating Model

```python
from keras.models import Sequential
from keras.layers import Dense
model = Sequential()
model.add(Dense(12, input_dim=28, activation='relu'))
model.add(Dense(8,activation='relu'))
model.add(Dense(1,activation='sigmoid'))
```

## Compile model

```python
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

## Fit the Model

```python
model.fit(X_standardized, y, validation_split=0.3, epochs=150, batch_size=10)
```

```python
scores = model.evaluate(X_standardized, y)
print("%s: %.2f%%" % (model.metrics_names[1], scores[1]*100))
```
